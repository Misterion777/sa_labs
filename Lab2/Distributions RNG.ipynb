```python
import numpy as np
from matplotlib import pyplot as plt
import math
plt.rcParams["figure.figsize"] = (16,9)
```

```python
N = int(1e6)
N
```

```python
def get_parameters(numbers):
    return f"Mean: {numbers.mean()}, Variance: {numbers.var()}, Std: {numbers.std()}"
```

## Uniform

```python
low = 6.5
high = 7
numbers = np.random.uniform(low=low, high = high, size=N)
```

```python
print(get_parameters(numbers))
```

```python
plt.hist(numbers, bins=20, density=True, rwidth=0.95)
plt.ylabel('Probability')
```

## Normal

```python
mu = 500.1
sigma = 40.1
numbers = np.random.normal(loc=mu, scale=sigma, size=N)
```

```python
print(get_parameters(numbers))
```

```python
plt.hist(numbers, bins=20, density=True, rwidth=0.95)
plt.ylabel('Probability')
```

## Exponential

```python
scale = 0.01 # lambda
numbers = np.random.exponential(scale=1/scale, size=N)
```

```python
print(get_parameters(numbers))
```

```python
plt.hist(numbers, bins=20, density=True, rwidth=0.95)
plt.ylabel('Probability')
```

## Gamma

```python
shape = 10.1 # n
scale = 10.1 #lambda
numbers = np.random.gamma(shape, scale=1/scale, size=N)
```

```python
print(get_parameters(numbers))
```

```python
plt.hist(numbers, bins=20, density=True, rwidth=0.95)
plt.ylabel('Probability')
```

## Triangular

```python
left = 0.1
right = 5.5
mode = right
numbers = np.random.triangular(left,mode,right, size=N)
```

```python
print(get_parameters(numbers))
```

```python
plt.hist(numbers, bins=20, density=True, rwidth=0.95)
plt.ylabel('Probability')
```

## Simpson

```python
left = 0.8
right = 5.1
mode = left + right / 2
numbers = np.random.triangular(left,mode,right, size=N)
```

```python
print(get_parameters(numbers))
```

```python
plt.hist(numbers, bins=20, density=True, rwidth=0.95)
plt.ylabel('Probability')
```
